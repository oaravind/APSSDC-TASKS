### Take one dataset and plot all graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bird.csv")
df
```

```python
plt.plot(df["feml"][:])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

```python
plt.scatter(df["huml"][:],df["humw"][:])
plt.xlabel('data of huml')
plt.ylabel('data of humw')
plt.title("Bird Dataset")
plt.show()
```

```python

from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df["type"] = le.fit_transform(df["type"])
df
```

```python
plt.hist(df["type"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

```python
plt.bar(df["type"],df["tarw"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

```python
plt.pie(df["huml"][:5],labels = df["humw"][:5],autopct = "% 1.1f%%")
plt.show()
```

```python
plt.boxplot(df["type"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```
